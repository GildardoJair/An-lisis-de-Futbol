## Medidas de Posición


Las medidas de posición dividen al conjunto de datos, previamente ordenado,
en grupos con determinada cantidad de observaciones. Las principales medidas
de posición son:

a. Cuartiles.

b. Percentiles.

### Cuartiles

Los cuartiles son valores que dividen al conjunto ordenado de observaciones en
cuatro partes iguales. Los cuartiles son denotados por Q1, Q2 y Q3.

```python
# Cuartiles
q1 = df['Age'].quantile(0.25)
q2 = df['Age'].quantile(0.50)  # o serie.median()
q3 = df['Age'].quantile(0.75)

print("Q1:", q1)
print("Q2 (mediana):", q2)
print("Q3:", q3)
```

### Percentiles
Los percentiles x_p(p = 1, 2, ..., 99) dividen al conjunto ordenado de datos en 100
partes iguales. De tal forma que p% de las observaciones son menores que el
percentil x_p y (100 – p)% de observaciones son al menos iguales a x_p. Los percentiles
son importantes en áreas como investigación de mercados pues permiten la
segmentación de estos.

```python
# Percentil 90
p90 = df['Age'].quantile(0.9)
print("Percentil 90:", p90)
```

### Boxplot

### 🧷 ¿Qué elementos ves en el boxplot?

**Caja (Box):**  
- Va desde el **Q1** (percentil 25%) hasta el **Q3** (percentil 75%).  
- La línea dentro de la caja representa la **mediana** (**Q2** o percentil 50%).

**Bigotes (Whiskers):**  
- Se extienden hasta los valores que están dentro de un rango de **1.5 veces el IQR** (rango intercuartílico).  
- **IQR = Q3 - Q1**

**Puntos fuera de los bigotes:**  
- Son considerados **valores atípicos** (*outliers*).


```python
plt.figure(figsize=(6, 4))
plt.boxplot(df['Age'], vert=True, patch_artist=True,
    whiskerprops=dict(color="white"),
    capprops=dict(color='white'),
    flierprops=dict(markerfacecolor="gray", markeredgecolor="white"),
    medianprops=dict(color='white'))
plt.title('Boxplot del conjunto de datos')
plt.xlabel('Valor')
plt.grid(True)
plt.show()
```

# Medidas de Dispersión

Las medidas de dispersión son valores que sirven para cuantificar la homogeneidad
(uniformidad, variabilidad) de los datos, es decir, sirven para medir
la proximidad que tienen los datos entre sí. Las medidas de dispersión
también son conocidas como medidas de variabilidad. Para el cálculo de algunas
de las medidas de dispersión se toma un punto de referencia que generalmente
es la media.

Siempre debe usarse, como mínimo, una medida de tendencia
central y una medida de dispersión para describir
el comportamiento de un conjunto de datos.

## Rango o Amplitud
El rango muestral es la medida de dispersión más simple y se denota por R; el
rango es la primera medida de dispersión que debe usarse porque permite conocer
el intervalo de variación de los datos. Las fórmulas para calcularlo se indican
a continuación.

```python
print('la edad  máxima es ',df['Age'].max())
print('la edad mínima es: ',df['Age'].min())
```

```python
rango = df['Age'].max() - df['Age'].min()
print(rango)
```

## Rango intercuantilo
La diferencia entre el tercer y el primer cuartil (Q3 –Q1) es conocido como el
rango (o amplitud) intercuartílico; dentro del mencionado rango se encuentra el
50 % central de las observaciones.

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1

print("IQR:", iqr)
```

## Varianza
La desviación de una observación con respecto a la media se define como: xi − x,
y puede asumir valores positivos o negativos dependiendo si el valor xi se encuentra
por encima o por debajo de la media.

```python
#Varianza muestral
varianza = df['Age'].var()
print(varianza)
```

```python
# Varianza poblacional
varianza_p = df['Age'].var(ddof=0)
print(varianza_p)
```

## Desviación Estándar
La desviación estándar muestral, denotada por S, cuantifica la dispersión de los
datos xi con respecto a la media.

```python
S = df['Age'].std()
print(S)
```

```python
S_Poblacional = df['Age'].std(ddof=0)
print(S_Poblacional)
```

```python
import numpy as np
# Acceder al ciclo de colores definido en axes.prop_cycle
color_ciclo = plt.rcParams['axes.prop_cycle'].by_key()['color']

# Elegir el segundo color del ciclo (índice 1)
color = color_ciclo[2]
columna = 'Age'
media = df[columna].mean()
std = df[columna].std()

plt.plot(df[columna], marker='o', linestyle='-', label='Datos')
plt.axhline(media,color = color, linestyle='--', label='Media')
plt.axhline(media + std, color='red', linestyle='--', label='+1 STD')
plt.axhline(media - std, color='red', linestyle='--', label='-1 STD')
plt.title(f"Dispersión de {columna}")
plt.legend()
plt.show()
```

## Coeficiente de Variación
El coeficiente de variación, denotado por C.V., cuantifica la dispersión relativa
que tienen los datos expresándola como el porcentaje de la desviación estándar
(S) con respecto al valor absoluto de la media (x), es decir, si x es el 100 %,
entonces el coeficiente de variación es el porcentaje de la desviación estándar
muestral con respecto a x.

La fórmula de cálculo de C.V. es la misma para datos agrupados y
no agrupados.
Las propiedades del coeficiente de variación son:

a. El coeficiente de variación no tiene unidad de medida.

b. El coeficiente de variación es útil para juzgar si un conjunto de datos
es homogéneo o heterogéneo. Para este fin, se deben utilizar valores
de referencia.

### Interpretación del Coeficiente de Variación (CV)

| CV (%)          | Interpretación                             |
|-----------------|---------------------------------------------|
| **Bajo (< 15%)**     | Poca variación relativa a la media         |
| **Moderado (15–30%)**| Variabilidad aceptable o moderada         |
| **Alto (> 30%)**     | Alta dispersión; datos menos consistentes |


```python
media_age = df['Age'].mean()
std_age = df['Age'].std()
cv_age = (std_age / media_age) * 100

print(f"Coeficiente de variación (Age): {cv_age:.2f}%")
```
