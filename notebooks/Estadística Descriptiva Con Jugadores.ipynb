## Medidas de Posici√≥n


Las medidas de posici√≥n dividen al conjunto de datos, previamente ordenado,
en grupos con determinada cantidad de observaciones. Las principales medidas
de posici√≥n son:

a. Cuartiles.

b. Percentiles.

### Cuartiles

Los cuartiles son valores que dividen al conjunto ordenado de observaciones en
cuatro partes iguales. Los cuartiles son denotados por Q1, Q2 y Q3.

```python
# Cuartiles
q1 = df['Age'].quantile(0.25)
q2 = df['Age'].quantile(0.50)  # o serie.median()
q3 = df['Age'].quantile(0.75)

print("Q1:", q1)
print("Q2 (mediana):", q2)
print("Q3:", q3)
```

### Percentiles
Los percentiles x_p(p = 1, 2, ..., 99) dividen al conjunto ordenado de datos en 100
partes iguales. De tal forma que p% de las observaciones son menores que el
percentil x_p y (100 ‚Äì p)% de observaciones son al menos iguales a x_p. Los percentiles
son importantes en √°reas como investigaci√≥n de mercados pues permiten la
segmentaci√≥n de estos.

```python
# Percentil 90
p90 = df['Age'].quantile(0.9)
print("Percentil 90:", p90)
```

### Boxplot

### üß∑ ¬øQu√© elementos ves en el boxplot?

**Caja (Box):**  
- Va desde el **Q1** (percentil 25%) hasta el **Q3** (percentil 75%).  
- La l√≠nea dentro de la caja representa la **mediana** (**Q2** o percentil 50%).

**Bigotes (Whiskers):**  
- Se extienden hasta los valores que est√°n dentro de un rango de **1.5 veces el IQR** (rango intercuart√≠lico).  
- **IQR = Q3 - Q1**

**Puntos fuera de los bigotes:**  
- Son considerados **valores at√≠picos** (*outliers*).


```python
plt.figure(figsize=(6, 4))
plt.boxplot(df['Age'], vert=True, patch_artist=True,
    whiskerprops=dict(color="white"),
    capprops=dict(color='white'),
    flierprops=dict(markerfacecolor="gray", markeredgecolor="white"),
    medianprops=dict(color='white'))
plt.title('Boxplot del conjunto de datos')
plt.xlabel('Valor')
plt.grid(True)
plt.show()
```

# Medidas de Dispersi√≥n

Las medidas de dispersi√≥n son valores que sirven para cuantificar la homogeneidad
(uniformidad, variabilidad) de los datos, es decir, sirven para medir
la proximidad que tienen los datos entre s√≠. Las medidas de dispersi√≥n
tambi√©n son conocidas como medidas de variabilidad. Para el c√°lculo de algunas
de las medidas de dispersi√≥n se toma un punto de referencia que generalmente
es la media.

Siempre debe usarse, como m√≠nimo, una medida de tendencia
central y una medida de dispersi√≥n para describir
el comportamiento de un conjunto de datos.

## Rango o Amplitud
El rango muestral es la medida de dispersi√≥n m√°s simple y se denota por R; el
rango es la primera medida de dispersi√≥n que debe usarse porque permite conocer
el intervalo de variaci√≥n de los datos. Las f√≥rmulas para calcularlo se indican
a continuaci√≥n.

```python
print('la edad  m√°xima es ',df['Age'].max())
print('la edad m√≠nima es: ',df['Age'].min())
```

```python
rango = df['Age'].max() - df['Age'].min()
print(rango)
```

## Rango intercuantilo
La diferencia entre el tercer y el primer cuartil (Q3 ‚ÄìQ1) es conocido como el
rango (o amplitud) intercuart√≠lico; dentro del mencionado rango se encuentra el
50 % central de las observaciones.

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1

print("IQR:", iqr)
```

## Varianza
La desviaci√≥n de una observaci√≥n con respecto a la media se define como: xi ‚àí x,
y puede asumir valores positivos o negativos dependiendo si el valor xi se encuentra
por encima o por debajo de la media.

```python
#Varianza muestral
varianza = df['Age'].var()
print(varianza)
```

```python
# Varianza poblacional
varianza_p = df['Age'].var(ddof=0)
print(varianza_p)
```

## Desviaci√≥n Est√°ndar
La desviaci√≥n est√°ndar muestral, denotada por S, cuantifica la dispersi√≥n de los
datos xi con respecto a la media.

```python
S = df['Age'].std()
print(S)
```

```python
S_Poblacional = df['Age'].std(ddof=0)
print(S_Poblacional)
```

```python
import numpy as np
# Acceder al ciclo de colores definido en axes.prop_cycle
color_ciclo = plt.rcParams['axes.prop_cycle'].by_key()['color']

# Elegir el segundo color del ciclo (√≠ndice 1)
color = color_ciclo[2]
columna = 'Age'
media = df[columna].mean()
std = df[columna].std()

plt.plot(df[columna], marker='o', linestyle='-', label='Datos')
plt.axhline(media,color = color, linestyle='--', label='Media')
plt.axhline(media + std, color='red', linestyle='--', label='+1 STD')
plt.axhline(media - std, color='red', linestyle='--', label='-1 STD')
plt.title(f"Dispersi√≥n de {columna}")
plt.legend()
plt.show()
```

## Coeficiente de Variaci√≥n
El coeficiente de variaci√≥n, denotado por C.V., cuantifica la dispersi√≥n relativa
que tienen los datos expres√°ndola como el porcentaje de la desviaci√≥n est√°ndar
(S) con respecto al valor absoluto de la media (x), es decir, si x es el 100 %,
entonces el coeficiente de variaci√≥n es el porcentaje de la desviaci√≥n est√°ndar
muestral con respecto a x.

La f√≥rmula de c√°lculo de C.V. es la misma para datos agrupados y
no agrupados.
Las propiedades del coeficiente de variaci√≥n son:

a. El coeficiente de variaci√≥n no tiene unidad de medida.

b. El coeficiente de variaci√≥n es √∫til para juzgar si un conjunto de datos
es homog√©neo o heterog√©neo. Para este fin, se deben utilizar valores
de referencia.

### Interpretaci√≥n del Coeficiente de Variaci√≥n (CV)

| CV (%)          | Interpretaci√≥n                             |
|-----------------|---------------------------------------------|
| **Bajo (< 15%)**     | Poca variaci√≥n relativa a la media         |
| **Moderado (15‚Äì30%)**| Variabilidad aceptable o moderada         |
| **Alto (> 30%)**     | Alta dispersi√≥n; datos menos consistentes |


```python
media_age = df['Age'].mean()
std_age = df['Age'].std()
cv_age = (std_age / media_age) * 100

print(f"Coeficiente de variaci√≥n (Age): {cv_age:.2f}%")
```
