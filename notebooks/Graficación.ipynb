# Usando el archivo neon_dark.mplstyle

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use(r"C:\Users\gilda\Desktop\Python\Proyecto Futbol\notebooks\neon_dark")
# Datos
x = np.linspace(0, 10, 500)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear figura
plt.figure(figsize=(10, 5))

# Dibujar con colores nen personalizados
plt.plot(x, y1, color='#FFA500', linewidth=2.5, label='Seno')     
plt.plot(x, y2, color='#00BFFF', linewidth=2.5, label='Coseno')   

# Personalizar
plt.title('Grfico con estilo Nen', fontsize=16, color='white')
plt.xlabel('Eje X', color='white')
plt.ylabel('Eje Y', color='white')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
